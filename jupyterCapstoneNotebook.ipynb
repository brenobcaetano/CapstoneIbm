```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
%matplotlib inline
```

```python
import sklearn.cluster as cluster
stdSca = std.copy()
```

```python
stdSca.drop(['state','state_po','Class'],axis=1,inplace=True)
```

```python
mms=MinMaxScaler()
mms.fit(stdSca)
data_transformed=mms.transform(stdSca)
```

```python
data_transformed=pd.DataFrame(data_transformed, columns=['mean','std'])
```

```python
data_transformed.head()
```

```python
data_transformed.plot(kind='scatter',x='std',y='mean')
plt.show()
```

```python
#Elbow method > minimize clusters
sum_square = []
K = range(1,15)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(data_transformed)
    sum_square.append(km.inertia_)
```

```python
plt.plot(K, sum_square, 'bx-')
plt.show()
```

```python
kmeans = cluster.KMeans(n_clusters=6, init='k-means++')
kmeans = kmeans.fit(std[['mean','std']])
```

```python
kmeans.cluster_centers_
```

```python
std['clusters'] = kmeans.labels_
```

```python
std.head()
```

```python
import seaborn as sns
```

```python
sns.scatterplot(x='std',y='mean',hue = 'clusters',data=std)
```

```python
std.sort_values(by='mean')
```

## Machine learning DF

```python
mcDF = df.copy()
```

```python
mcDF['d%'] = mcDF['demVotes']/mcDF['totalvotes']
mcDF['r%'] = mcDF['repVotes']/mcDF['totalvotes']
```

```python
mcDF.head()
```

```python
mcDF = mcDF.groupby(by=['state']).sum()
```

```python
mcDF.head()
```

```python
mcDF = mcDF.reset_index()
```

```python
mcDF.head()
```

```python
mcDF.drop(['index','year','state','totalvotes','repPercent','demPercent','otherVotes','repVotes','demVotes'],axis=1,inplace=True)
```

```python
mcDF.head()
```

```python
mms=MinMaxScaler()
mms.fit(mcDF)
data_transformed=mms.transform(mcDF)
```

```python
data_transformed=pd.DataFrame(data_transformed, columns=['d%','r%'])
```

```python
data_transformed.head()
```

```python
data_transformed.plot(kind='scatter',x='d%',y='r%')
plt.show()
```

```python
#Elbow method > minimize clusters
sum_square = []
K = range(1,15)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(data_transformed)
    sum_square.append(km.inertia_)
```

```python
plt.plot(K, sum_square, 'bx-')
plt.show()
```

```python
kmeans = cluster.KMeans(n_clusters=3, init='k-means++')
kmeans = kmeans.fit(mcDF[['d%','r%']])
```

```python
kmeans.cluster_centers_
```

```python
std['clusters'] = kmeans.labels_
```

```python
std['Democrat %'] = mcDF['d%']
```

```python
std['Republican %'] = mcDF['r%']
```
